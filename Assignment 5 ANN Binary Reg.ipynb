```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("Bank_Personal_Loan_Modelling.csv")
```

```python
dataset.drop(columns=["ID","ZIP Code"],inplace=True)
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,[0,1,2,3,4,5,6,8,9,10,11]].values
y=dataset.iloc[:,7].values
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) #test_size=0.2 is 20% testing data
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #used to initialize
from tensorflow.keras.layers import Dense #used to add layers
```

```python
ann=Sequential()
```

```python
ann.add(Dense(11,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(22,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(1,kernel_initializer ="random_uniform",activation="sigmoid"))
```

```python
ann.compile("adam","binary_crossentropy",metrics=["accuracy"]) #nick name for batch gradient is adam.
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=30)
```

```python
ypred=ann.predict(x_test)
```
